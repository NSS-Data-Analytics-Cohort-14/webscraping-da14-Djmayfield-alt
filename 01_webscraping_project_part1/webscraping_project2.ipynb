```python
jobs = soup.find(id="ResultsContainer")
jib_jab = jobs.find_all("div", class_="card-content")
```

### c. Finally, extract the companies, locations, and posting dates for each job. For example, the first job has a company of "Payne, Roberts and Davis", a location of "Stewartbury, AA", and a posting date of "2021-04-08". Ensure that the text that you extract is clean, meaning no extra spaces or other characters at the beginning or end. 

```python
for job_card in jib_jab:
    title = job_card.find('h2', class_='title').text.strip()
    company = job_card.find('h3', class_='company').text.strip()
    location = job_card.find('p', class_='location').text.strip()
    date = job_card.find('time').text.strip()
    print(title)
    print(company)
    print(location)
    print(date)
    print()
```

### d. Take the lists that you have created and combine them into a pandas DataFrame. 

```python
title = []
company = []
location = []
date = []

for job_card in jib_jab:
    title1 = job_card.find('h2', class_='title').text.strip()
    company1 = job_card.find('h3', class_='company').text.strip()
    location1 = job_card.find('p', class_='location').text.strip()
    date1 = job_card.find('time').text.strip()
    title.append(title1)
    company.append(company1)
    location.append(location1)
    date.append(date1)


data = {
    'Title': title,
    'Company': company,
    'Location': location,
    'Date Posted': date
}

df=pd.DataFrame(data)

df
```
